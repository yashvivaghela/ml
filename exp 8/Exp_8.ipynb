```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/iris.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
X = df[["sepal.length",	"sepal.width","petal.length","petal.width"]]
y = df['variety']
```

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
scaled_x = ss.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(scaled_x,y)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.decomposition import PCA
pcam=PCA(n_components=3)
x_pca=pcam.fit_transform(scaled_x)

print("Eigen Vectors =\n",pcam.components_)
print("\nEigen Values =\n",pcam.explained_variance_)
```

```python
print(x_pca)
```

```python
from sklearn.model_selection import train_test_split
x_pca_train,x_pca_test, y_train, y_test = train_test_split(x_pca,y)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_pca_train,y_train)
model.score(x_pca_test,y_test)
```
